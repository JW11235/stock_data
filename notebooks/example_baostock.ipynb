### K线信息

```python
stock_code = 'sh.600000'
fields = 'date,code,open,high,low,close,preclose,volume,amount,adjustflag,\
         turn,tradestatus,pctChg,isST'
start_date = '2017-07-01'
end_date = '2017-12-31'
    
rs = bs.query_history_k_data_plus(
                                    stock_code, 
                                    fields, 
                                    start_date=start_date, 
                                    end_date=end_date,
                                    frequency='d', 
                                    adjustflag='3'
)
```

```python
data_list = []
while (rs.error_code == '0') & rs.next():
    data_list.append(rs.get_row_data())
    
result = pd.DataFrame(data_list, columns=rs.fields)
```

```python
result
```

```python
result.iloc[:, 2:-1] = result.iloc[:, 2:-1].applymap(lambda x: float(x))
result['code'] = result['code'].apply(lambda x: x[3:] + '.' + x[0:2])
```
